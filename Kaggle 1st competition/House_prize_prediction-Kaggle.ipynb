```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = pd.read_csv('train.csv').set_index("Id")
df2 = pd.read_csv('test.csv').set_index('Id')
data1 = df1.iloc[:,:-1].values
data2 = df2.iloc[:,:].values
df3 = np.concatenate((data1,data2),axis=0) 
df = pd.DataFrame(data=df3)
```

```python
pd.set_option("max_columns",None)
df.columns = df2.columns
```

```python
for col in df.columns:
    if df[col].isnull().sum()>650:
        print(col)
        del df[col]
```

```python
column = df.columns
typ = type(df['Street'][0])
column = [col for col in column if not type(df[col][0])== typ]
int_df = df[np.array(column)].copy()
int_df.head()
```

```python
column = np.array(int_df.columns)
print(column)
for col in column:
    if int_df[col].isnull().sum()<5:
        int_df[col].fillna(int_df[col].mean(), inplace=True)
int_df.head()
del int_df['LotFrontage']
```

```python
int_df["GarageYrBlt"] = int_df["GarageYrBlt"].fillna(int_df["GarageYrBlt"].mean())
del int_df["MasVnrArea"]
int_df.isnull().sum()
```

```python
df_new = int_df.dropna().astype('int').copy()
```

```python
df_new
```

```python
column = df.columns
typ = type(df['Street'][0])
column = [col for col in column if type(df[col][0])== typ]
print(column)
str_df = df[column]
str_df.head()
```

```python
str_df["BsmtQual"] = str_df["BsmtQual"].fillna(str_df["BsmtQual"].mode()[0])
del str_df["BsmtCond"]
str_df["BsmtExposure"] = str_df["BsmtExposure"].fillna(str_df["BsmtExposure"].mode()[0])
str_df["BsmtFinType1"] = str_df["BsmtFinType1"].fillna(str_df["BsmtFinType1"].mode()[0])
del str_df["BsmtFinType2"]
```

```python
str_df.isnull().sum()
```

```python
str_df["MSZoning"] = str_df["MSZoning"].fillna("RH")
del str_df["Utilities"]
del str_df["Exterior2nd"]
str_df["Exterior1st"] = str_df["Exterior1st"].fillna("BrkComm")
str_df["Exterior1st"] = str_df["Exterior1st"].replace(['Stone', 'ImStucc','CBlock','AsphShn'],['BrkComm', 'BrkComm', 'BrkComm', 'BrkComm'])
str_df["Exterior1st"].value_counts()
del str_df["Electrical"]
str_df["MasVnrType"] = str_df["MasVnrType"].fillna("BrkCmn")
str_df["KitchenQual"] = str_df["KitchenQual"].fillna("Fa")
del str_df["Functional"]
del str_df["GarageQual"]
del str_df["GarageCond"]
del str_df["SaleType"]
str_df["GarageType"] = str_df["GarageType"].fillna(str_df["GarageType"].mode()[0])
str_df["GarageFinish"] = str_df["GarageFinish"].fillna("RF")
```

```python
str_df.isnull().sum()
```

```python
str_columns = str_df.columns
for col in str_columns:
    vect = pd.get_dummies(df[col], prefix=col+"_")
    del str_df[col]
    str_df = pd.concat([str_df,vect], axis = 1)
    
str_df.shape
```

```python
df = pd.concat([int_df,str_df],axis=1)
```

```python
df.corr()
```

```python
X_train = df.iloc[:1460,:].values
X_test = df.iloc[1460:,:].values
y_train = df1.iloc[:,-1].values
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 10)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 15, random_state = 0)
regressor.fit(X_train, y_train)
#  Random Forest regression is the best till now
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
pred = pd.DataFrame(y_pred)
sub_df = pd.read_csv('sample_submission.csv')
dataset = pd.concat([sub_df['Id'],pred],axis=1)
dataset.columns = sub_df.columns
dataset.to_csv('sample_submission.csv',index=False)
```
